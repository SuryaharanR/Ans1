```python
import numpy as np
import matplotlib.pyplot as plt

data_1 = np.loadtxt('Data_1.txt')
data_2 = np.loadtxt('Data_2.txt')

def find_peaks(data):
    maxima = []
    minima = []
    for i in range(1, len(data) - 1):
        if data[i] > data[i - 1] and data[i] > data[i + 1]:
            maxima.append(i)
        elif data[i] < data[i - 1] and data[i] < data[i + 1]:
            minima.append(i)
    return maxima, minima

maxima_1, minima_1 = find_peaks(data_1)
maxima_2, minima_2 = find_peaks(data_2)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(data_1, label='Data_1')
ax1.scatter(maxima_1, data_1[maxima_1], color='red', label='Maxima', marker='o')
ax1.scatter(minima_1, data_1[minima_1], color='blue', label='Minima', marker='o')
ax1.set_title('Data_1')
ax1.legend()

ax2.plot(data_2, label='Data_2')
ax2.scatter(maxima_2, data_2[maxima_2], color='red', label='Maxima', marker='o')
ax2.scatter(minima_2, data_2[minima_2], color='blue', label='Minima', marker='o')
ax2.set_title('Data_2')
ax2.legend()

plt.tight_layout()
plt.show()
```
